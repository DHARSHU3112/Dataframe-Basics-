```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the chart
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Chart')

# Show the chart
plt.show()


```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the chart
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Chart')

# Show the chart
plt.show()

# Customizing the line color, style, and markers
plt.plot(x, y, color='green', linestyle='--', marker='o')

# Add grid
plt.grid(True)
```

```python
                                                   #Bar Chart
# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the chart
plt.bar(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Bar Chart')

# Show the chart
plt.show()


```

```python
                                                                     # Scatter Chart
# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the chart
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Chart')

# Show the chart
plt.show()
```

```python
                                                     ## Histogram
# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the chart
plt.hist(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Histogram')

# Show the chart
plt.show()
```
